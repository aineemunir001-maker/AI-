```python
#TASK 6
text = "Python is simple yet powerful. Python is easy to learn."
word_freq = {}
words = text.split()

for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

print(word_freq)
```
